```python
df2['diagnosis'] = df['diagnosis'].apply(lambda x: 1 if x == 'M' else 0)
```

```python
df2.head()
```

```python
df2.drop(columns=['id'], axis=1, inplace=True)
```

```python
df2.corr().abs()
```

```python
alvo = df2['diagnosis'].values
alvo
```

```python
df2.shape
```

```python
alvo.shape
```

```python
previsores = df2.iloc[:, 1:].values
previsores
```

```python
previsores.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
previsores_esc = StandardScaler().fit_transform(previsores)
previsores_esc
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(previsores, alvo, test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random = RandomForestClassifier(n_estimators=120, criterion='entropy', random_state = 2, max_depth=6)
random.fit(x_treino, y_treino)
```

```python
previsoes_random = random.predict(x_teste)
previsoes_random
```

```python
y_teste
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_random) * 100.0))
```

```python
confusion_matrix(y_teste, previsoes_random)
```

```python
print(classification_report(y_teste, previsoes_random))
```

```python
previsoes_treino = random.predict(x_treino)
previsoes_treino
```

```python
accuracy_score(y_treino, previsoes_treino)
```

```python
confusion_matrix(y_treino, previsoes_treino)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
kfold = KFold(n_splits = 30, shuffle=True, random_state = 5)
```

```python
# Criando o modelo
modelo = RandomForestClassifier(n_estimators=120, criterion='entropy', random_state = 2, max_depth=6)
resultado = cross_val_score(modelo, previsores, alvo, cv = kfold)

# Usamos a média e o desvio padrão
print("Acurácia Média: %.2f%%" % (resultado.mean() * 100.0))
```
